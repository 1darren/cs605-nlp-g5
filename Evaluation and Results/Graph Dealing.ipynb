```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Change to the directory where your CSV file is located
os.chdir(os.path.expanduser('~/Downloads/Inference'))

# Read the CSV file with scores
file_path = 'questions_answers_with_scores-cleaned-gpt4o-complete.csv'  # Replace with your actual file path
df = pd.read_csv(file_path)

# Ensure the CSV has the correct columns
score_columns = [col for col in df.columns if col.endswith('_Score')]

# Calculate averages in chunks of 10
chunk_size = 10
average_df = pd.DataFrame()

for col in score_columns:
    averages = [df[col][i:i + chunk_size].mean() for i in range(0, len(df), chunk_size)]
    average_df[col] = averages

# Plot the scores for each answer
plt.figure(figsize=(12, 8))

# Define x_values based on chunk_size, representing the begin-point of each chunk
x_values = [i for i in range(0, len(df), chunk_size)]

# Define x_values based on chunk_size, representing the mid-point of each chunk
# x_values = [i + chunk_size/2 for i in range(0, len(df), chunk_size)]

for col in average_df.columns:
    plt.plot(x_values, average_df[col], label=col)

plt.xlabel('Answers')
plt.ylabel('Score')
plt.title('Scores for Each Answer from GPT4o')
plt.grid(True)

# Set x-axis labels from 0 to 100 in increments of 20
x_ticks = list(range(0, 100, 10))
plt.xticks(x_ticks)

# Move the legend outside the plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.85))  # Position the legend outside the plot

plt.tight_layout()  # Adjust layout to make room for the legend
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# Change to the directory where your CSV file is located
os.chdir(os.path.expanduser('~/Downloads/Inference/Final'))

# Read the CSV file with scores
file_path = 'questions_answers_with_scores_cleaned-final score.csv'  # Replace with your actual file path
df = pd.read_csv(file_path)

# Ensure the CSV has the correct columns
score_columns = [col for col in df.columns if col.endswith('_Score')]

# Calculate the average score for each column
average_scores = {col: df[col].mean() for col in score_columns}

# Convert the dictionary to a DataFrame for plotting
average_scores_df = pd.DataFrame(list(average_scores.items()), columns=['Score Type', 'Average Score'])

# Plot the average scores as a bar graph with different colors
plt.figure(figsize=(12, 8))

# Generate a color for each bar
colors = plt.cm.viridis_r(np.linspace(0, 1, len(score_columns)))

# Define x positions for bars
x_positions = np.arange(len(average_scores_df))

plt.bar(x_positions, average_scores_df['Average Score'], color=colors)

#plt.xlabel('Score Type')
plt.ylabel('Average Score')
plt.title('Comparison of Average Scores for 100 Answers from GPT4o')
plt.xticks(x_positions-0.5, average_scores_df['Score Type'], rotation=45, fontsize=12)  # Align labels with bars
plt.grid(True)

# Adjust layout to make room for the labels
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# Change to the directory where your CSV files are located
os.chdir(os.path.expanduser('~/Downloads/Inference'))

# List of CSV file paths
csv_files = [
    'questions_answers_with_scores_cleaned-c.csv',
    'questions_answers_with_scores_cleaned-ca.csv',
    'questions_answers_with_scores_cleaned-s.csv',
    'questions_answers_with_scores_cleaned-v.csv'
]

# Initialize a list to store DataFrames of total scores
total_scores_list = []

# Read each CSV file and calculate total scores
for file in csv_files:
    df = pd.read_csv(file)
    # Ensure the CSV has the correct columns
    score_columns = [col for col in df.columns if col.endswith('_Score')]
    # Convert columns to numeric, setting errors='coerce' will replace non-numeric values with NaN
    df[score_columns] = df[score_columns].apply(pd.to_numeric, errors='coerce')
    # Fill NaN values with 0
    df[score_columns] = df[score_columns].fillna(0)
    # Cap scores at a maximum reasonable value, e.g., 100
    df[score_columns] = df[score_columns].applymap(lambda x: min(x, 100))
    
    # Print intermediate results for verification
    print(f"Processed data from file {file}:")
    print(df[score_columns].head())
    
    total_scores = {col: df[col].sum() for col in score_columns}
    print(f"Total scores from file {file}: {total_scores}")
    
    # Convert the total scores to a DataFrame and append to the list
    total_scores_df = pd.DataFrame(total_scores, index=[file])
    total_scores_list.append(total_scores_df)

# Concatenate all DataFrames in the list
combined_total_scores_df = pd.concat(total_scores_list)

# Transpose the DataFrame to have score types as rows and CSV files as columns
combined_total_scores_df = combined_total_scores_df.T
combined_total_scores_df.columns = ['Vocabulary', 'Creativity', 'Sentiment', 'Character Accuracy']
combined_total_scores_df['Score Type'] = combined_total_scores_df.index

# Average the total scores with 100
combined_total_scores_df.iloc[:, :-1] = combined_total_scores_df.iloc[:, :-1].div(100)

# Print combined total scores for verification
print("Combined total scores (averaged with 100):")
print(combined_total_scores_df)

# Plot the total scores as a bar graph with different colors for each CSV
plt.figure(figsize=(16, 9))

# Define x positions for groups of bars
x_positions = np.arange(len(combined_total_scores_df))

# Define the width of each bar
bar_width = 0.15

# Colors for each CSV file
colors = ['#92A8D1', '#034F84', '#B565A7', '#6B5B95']

# Plot each CSV file's scores
for i, col in enumerate(combined_total_scores_df.columns[:-1]):
    plt.bar(x_positions + i * bar_width, combined_total_scores_df[col], width=bar_width, color=colors[i], label=col)

#plt.xlabel('Score Type')
plt.ylabel('Average Score')
plt.title('Average Scores for 100 Answers from GPT4o Across 4 CSVs')
# Define the font properties
font_properties = {'family': 'serif', 'weight': 'bold'}

# Set Y label with specific font properties
plt.ylabel('Average Score', fontdict={**font_properties, 'size': 14})

# Set title with specific font properties (optional)
plt.title('Average Scores for 100 Answers from GPT4o across 4 Aspects', fontdict={**font_properties, 'size': 16})

# Set X ticks with rotation and specific font size
plt.xticks(x_positions - 0.5 + bar_width * 1, combined_total_scores_df['Score Type'], rotation=45, fontsize=12)  # Align labels with bars

# Customize the legend font properties
plt.legend(prop={**font_properties, 'size': 12})  # Example: size 12, bold, serif font
plt.grid(False)

# Adjust layout to make room for the labels
plt.tight_layout()
plt.show()
```
