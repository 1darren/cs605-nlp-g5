```python
import pandas as pd
import matplotlib.pyplot as plt

os.chdir(os.path.expanduser('~/Downloads'))
# Read the CSV files
file1 = 'filled_answers_with prompt.csv'
file2 = 'filled_answers _raw.csv'

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# Ensure both files have the same length and the column exists
if len(df1) != len(df2):
    raise ValueError("CSV files must have the same number of rows.")

if 'Answer-1_Ricky_Score' not in df1.columns or 'Answer-1_Ricky_Score' not in df2.columns:
    raise ValueError("Column 'Answer-1_Ricky_Score' must exist in both files.")

# Extract the 'Answer-1_Ricky_Score' columns
scores1 = df1['Answer-1_Ricky_Score']
scores2 = df2['Answer-1_Ricky_Score']

# Filter the scores to only include values between 0.9 and 1.0
filtered_indices = (scores1.between(0.9, 1.0)) & (scores2.between(0.9, 1.0))
scores1_filtered = scores1[filtered_indices]
scores2_filtered = scores2[filtered_indices]

# Plot the scores from both files
plt.figure(figsize=(10, 6))

plt.plot(scores1_filtered.values, label='GPT-RAG-Prompt', color='blue')
plt.plot(scores2_filtered.values, label='GPT-RAG-Raw', color='red')

plt.fill_between(range(len(scores1_filtered)), scores1_filtered, color='blue', alpha=0.1)
plt.fill_between(range(len(scores2_filtered)), scores2_filtered, color='red', alpha=0.1)

plt.ylim(0.9, 1.0)
plt.xlabel('Index')
plt.ylabel('Answer-1_Ricky_Score')
plt.title('Answer-1_Ricky_Score Comparison Between Two CSV Files (0.9 to 1.0)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# Change to the directory where your CSV files are located
os.chdir(os.path.expanduser('~/Downloads/Inference'))

# List of CSV file paths
csv_files = [
    'questions_answers_with_scores_cleaned-c.csv',
    'questions_answers_with_scores_cleaned-ca.csv',
    'questions_answers_with_scores_cleaned-s.csv',
    'questions_answers_with_scores_cleaned-v.csv'
]

# Initialize a list to store DataFrames of total scores
total_scores_list = []

# Read each CSV file and calculate total scores
for file in csv_files:
    df = pd.read_csv(file)
    # Ensure the CSV has the correct columns
    score_columns = [col for col in df.columns if col.endswith('_Score')]
    total_scores = {col: df[col].sum() for col in score_columns}
    # Convert the total scores to a DataFrame and append to the list
    total_scores_df = pd.DataFrame(total_scores, index=[file])
    total_scores_list.append(total_scores_df)

# Concatenate all DataFrames in the list
combined_total_scores_df = pd.concat(total_scores_list)

# Transpose the DataFrame to have score types as rows and CSV files as columns
combined_total_scores_df = combined_total_scores_df.T
combined_total_scores_df.columns = ['CSV 1', 'CSV 2', 'CSV 3', 'CSV 4']
combined_total_scores_df['Score Type'] = combined_total_scores_df.index

# Plot the total scores as a bar graph with different colors for each CSV
plt.figure(figsize=(12, 8))

# Define x positions for groups of bars
x_positions = np.arange(len(combined_total_scores_df))

# Define the width of each bar
bar_width = 0.2

# Colors for each CSV file
colors = ['blue', 'green', 'red', 'purple']

# Plot each CSV file's scores
for i, col in enumerate(combined_total_scores_df.columns[:-1]):
    plt.bar(x_positions + i * bar_width, combined_total_scores_df[col], width=bar_width, color=colors[i], label=col)

plt.xlabel('Score Type')
plt.ylabel('Total Score')
plt.title('Total Scores for 100 Answers from GPT4o Across 4 CSVs')
plt.xticks(x_positions + bar_width * 1.5, combined_total_scores_df['Score Type'], rotation=45, fontsize=12)  # Align labels with bars
plt.legend()
plt.grid(True)

# Adjust layout to make room for the labels
plt.tight_layout()
plt.show()
```
